# Azure Data Studio

> Azure Data Studio \[[GitHub](https://github.com/Microsoft/azuredatastudio)\] \[[docs](https://docs.microsoft.com/en-us/sql/azure-data-studio/?view=sql-server-ver15)\] is a data management tool that enables you to work with SQL Server, Azure SQL DB and SQL DW from Windows, macOS and Linux.

Leading conveniences of Azure Data Studio:

- connects to SQL Azure as well as local SQL Server
- cross-platform (unlike [SSMS](https://docs.microsoft.com/en-us/sql/ssms/sql-server-management-studio-ssms?view=sql-server-ver15))
- probably manages connections (securely?) better out of the box than Visual Studio Code
- supports Jupyter Notebooks without forcing users to learn Markdown

Leading inconveniences of Azure Data Studio:

- does not save output (that can be displayed outside of Azure Data Studio) from queries by default
- no dark mode
- text entry not as convenient as the current .NET Interactive UX

## Connecting to local SQL Server

After adding a connection to the local SQL Server in the Studio, a simple `SELECT` projection works as expected:

```sql
SELECT *
FROM [master].[sys].[columns]
WHERE [name] = 'id'
```

However the tabular output (above) does not display in other Jupyter-Notebook (`*.ipynb`) readers.
