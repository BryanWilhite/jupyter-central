# mermaid CLI

When [mermaid](https://mermaid-js.github.io/) diagrams get too large for a screen shot, we can use the mermaid CLI to convert a `*.mmd` file to various image formats.

The [mermaid CLI](https://github.com/mermaid-js/mermaid-cli) is installed as a local `npm` package:

```shell
npm install @mermaid-js/mermaid-cli
```

The mermaid CLI is invoked with `mmdc`:

````python
! npx mmdc --version
````

## `mmdc --help`

````python
! npx mmdc --help
````

## PNG output

````python
! npx mmdc \
    --input ./songhay-app-services.mmd \
    --output ./songhay-app-services.png \
    --theme forest \
    --width 1080

from IPython import display

image = display.Image(filename='./songhay-app-services.png')

image
````

````python
import os

os.path.getsize(image.filename)
````

## SVG output

````python
! npx mmdc \
    --input ./songhay-app-services.mmd \
    --output ./songhay-app-services.svg \
    --theme forest \
    --width 1080

from IPython.display import SVG

image = SVG(filename='./songhay-app-services.svg')

image
````
