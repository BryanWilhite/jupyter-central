This notebook documents months 11 and 12 for 2019 and month 1 for 2020.

```python
%run ./GreenlightUtility.py

averages = []
```

```python
starting_balance = 542.83

df_csv = GreenlightUtility.get_csv_data_frame('./data/2019-11.csv')
df_csv
```

```python
df_balances = GreenlightUtility.get_balance_data_frame(df_csv, starting_balance)
df_balances
```

```python
first_day_of_month = GreenlightUtility.get_first_day_of_month(2019, 11)

df_daily_balance = GreenlightUtility.get_daily_balance_data_frame(
    first_day_of_month,
    starting_balance,
    df_eod_balances)

averages.append((first_day_of_month.strftime('%Y-%m'), df_daily_balance.mean().values[0]))
```
