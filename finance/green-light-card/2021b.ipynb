# 2021 Average-Daily-Balance Calculations

This notebook documents months 2 through 4 in 2021.

```python
%run ./GreenlightUtility.py

averages = []
```

```python
starting_balance = 917.67

df_csv = GreenlightUtility.get_csv_data_frame('./data/2021-02.csv')
df_csv
```

```python
df_balances = GreenlightUtility.get_balance_data_frame(df_csv, starting_balance)
df_balances
```

```python
df_eod_balances = GreenlightUtility.get_balance_eod_data_frame(df_balances)
df_eod_balances
```

```python
first_day_of_month = GreenlightUtility.get_first_day_of_month(2021, 2)

df_daily_balance = GreenlightUtility.get_daily_balance_data_frame(
    first_day_of_month,
    starting_balance,
    df_eod_balances)

averages.append((first_day_of_month.strftime('%Y-%m'), df_daily_balance.mean().values[0]))
```

```python
df_eod_balances.tail(1)
```
