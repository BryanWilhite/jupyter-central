# Songhay Publications Presentation data legacy formats

The original Presentation data formats were in XML, generated from SQL Server data. These XML files were saved in a Presentation folder of the following form:

```console
<presentation-key>
|
\jpg
|  |___background.jpg
|
\xml
   |___client.xml
   |___<presentation-key>.xml
```

The `<presentation-key>.xml` file was designed to be edited on the fly in [Microsoft InfoPath](https://en.wikipedia.org/wiki/Microsoft_InfoPath) (see “the remaining Presentation JSON elements are ‘loose’ conventions” below).

XML was welcomed on the Web via [Macromedia/Adobe Flash](https://en.wikipedia.org/wiki/Adobe_Flash) plugins, running in the browser which could parse XML on the client side and render the Presentation.

When Flash was replaced with Angular JS and HTML5 at the turn of the century, this XML format was brute-force translated to JSON and looks like this:

JSON properties like `@ClientId` that begin with the `@` symbol indicate that they were translated from XML attributes—which betrays the brute-force nature of the translation. The JSON properties with the name `#text` indicate translation from XML text nodes.


## legacy Presentation JSON elements came from the GenericWeb schema



`Songhay.Publications` absorbs the `GenericWeb` schema, featuring [the Segment, Document and Fragment](https://github.com/BryanWilhite/Songhay.Publications#genericweb-models-and-extensions) which was historically enforced by SQL Server. The table below summarizes the `GenericWeb` sourced legacy JSON properties:

| JSON property | Publications property | remarks |
|- |- |-
| `Presentation.@ClientId` | `Segment.ClientId` | The client-side ID was a way to obscure internal SQL Server IDs; the need for this started with InfoPath. |
| `Presentation.@ModDate` | `Segment.ModificationDate` | |
| `Presentation.@PubDate` | `Segment.InceptDate` | |
| `Presentation.@Template` | `Document.TemplateId` | …where `Document.IsRoot == true`<sup>1</sup>; the XSLT Template was used to take SQL Server XML and transform it into XHTML. The `TemplateId` was used to select the XSL template. |
| `Presentation.@IntroUri` | `Document.Path` and `Document.FileName` | …where `Document.IsRoot == true`<sup>2</sup> |
| `Presentation.Title` | `Segment.SegmentName` | |
| `Presentation.Description.#text` | `Fragment.ItemText` | The `Fragment` with the Description text was associated with the root Document.<sup>3</sup> |
| `Presentation.Credits.#text` | `Fragment.ItemText` | No need to describe how this `Fragment` data was associated: not needed going forward. |
| `Presentation.ItemGroup.Item[]` | `Document[]` | The `Item` array has objects that map to Documents associated with the parent Segment backing the Presentation. |

<sup>1</sup><small>The _root_ Document represents the introductory content of the presentation.</small>

<sup>2</sup><small>Separating Document path from file name was probably useful for grouping hit counts below the Segment level.</small>

<sup>3</sup><small>Going forward, `Fragment` data should be replaced by Markdown Document data.</small>


## the remaining Presentation JSON elements are ‘loose’ conventions


The historical role InfoPath played was to support the entry of the remaining presentation elements (a temporary move that lasted for years) which were considered outside of the GenericWeb schema.<sup>1</sup>

| JSON property | remarks |
|- |-
| `Presentation.LayoutMetadata` | originally translated from the conventional `client.xml` file used by Flash and expanded to support CSS |
| `Presentation.Copyright.@Name` | ad hoc legal info |
| `Presentation.Copyright.@Year` | ad hoc legal info |

<sup>1</sup><small>To better support InfoPath, a `WebPresentation.xsd` XML schema was written and published.</small>
